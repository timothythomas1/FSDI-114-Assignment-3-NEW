```python
class QNode:
    def __init__(self, data):
        self.data = data
        self.prev = None

class Queue:
    def __init__(self):
        # Class attributes 
        self.front = None        
        
    # Class methods
    def enqueue(self, item): 
        if not self.front:  # If there is no Node instance at self.front
            self.front = QNode(item)
        else: 
            current = self.front
            while current.prev:
                current = current.prev
            current.prev = QNode(item)
        
    def dequeue(self):
        if self.front:
            current = self.front
            node_before = self.front.prev
            if not node_before:
                self.front = None
                return current.data
            else:
                self.front = node_before
                return current.data
        raise IndexError("Queue is empty") # This is here so that other developers can use your exception handling
        # return -1 <-- this was how it was done back in the day with C++
        
    
    def queue_size(self):
        count = 0
        current = self.front
        while current:
            count += 1
            current = current.prev
        return count
    
    def __str__(self): #Double underscore method ('dunder method')
        out_str = "["
        if self.front:
            current = self.front
            out_str += "%s" % current.data
            while current:                    # While current is NOT None.
                current = current.prev
                if current:
                    out_str += ", %s" % current.data
        out_str += "]"
        return out_str
    
#     def is 
```

```python
# Create a data structure that stores data items in sequence using pointers with methods PUSH, POP, PEEK and isEmpty.
class SLNode:
    def __init__(self, data): # <--  
        self.data = data
        self.next = None        

class Stack:
    def __init__(self):
        self.head = None
        
    def push(self, data): # Push to TOP of stack
        NewNode = SLNode(data) # Initialize NewNode to be appended
        if self.head is None:        # CASE 1: Consider when the SLL() has no Node()s.
            self.head = NewNode # --> Create an empty instance of the Node class and assign it to SLL's head attribute
            return
          # CASE 2: Consider when the SinglyLinkedList has a Node()-head-obj.
          # --> First, create a placeholder 'pointer' that points to this Node()-head-obj.
        temp_head_grab = self.head
        self.head = NewNode
        self.head.next = temp_head_grab

    def pop(self): # Remove from TOP of stack
        current = self.head
        if current is None:
            return "Stack is empty. Nothing to remove."
        next_head_node = current.next
        self.head = next_head_node
        return print(f'Popped "{current.data}" from the top of the stack.')
    #         raise IndexError("Stack is empty") # This is here so that other developers can use your exception handling
    #         # return -1 <-- this was how it was done back in the day with C++

    def stack_size(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count
    
    def peek(self):
        if self.head is None:
            return "The stack is empty."
        return (f'Peeking shows "{self.head.data}" is at the top of this stack.')
    
    def __str__(self): #Double underscore method ('dunder method')
        out_str = "["
        if self.head:
            pointer = self.head
            out_str += "%s" % pointer.data
            while pointer:                    # While current is NOT None.
                pointer = pointer.next
                if pointer:
                    out_str += ", %s" % pointer.data
        out_str += "]"
        return out_str
        
        
```

```python
def push_helper2(mylist=[1,"1", "op",3.76,"Hola!",[],{}]):
    newSLL = Stack()
    for value in mylist:
        newSLL.push(value)
        print(f'Pushed {type(value)} "{value}" to my SLL')
        print(f'Current Stack size {newSLL.stack_size()}')
        print(newSLL,  "\n")
        
    while newSLL.head:
        newSLL.pop()
        print(newSLL.peek())
        print(f'Current Stack size {newSLL.stack_size()}')
        print(newSLL,  "\n")
        

push_helper2()
```
